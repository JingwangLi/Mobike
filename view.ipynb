```python
import numpy as np
import pandas as pd
import os
import folium
data = pd.read_csv("E:\\graduate\\Mobike\\data\\20180901-115849-027.csv")
data.columns = ["time", "ID", "lng","lat"]
#max_lng, min_lng = max(data.lng), min(data.lng)
#max_lat, min_lat = max(data.lat), min(data.lat)
max_lng, max_lat = 114.462789,30.53242
min_lng, min_lat = 114.377558,30.48774
dlng, dlat = (max_lng - min_lng)/100, (max_lat - min_lat)/100
#print("dlng: %f dlat: %f" % (dlng, dlat))

Map = np.zeros((100,100), dtype = np.int)

def count(df):
    x = int((df['lng'] - min_lng) / dlng) - 1
    y = int((df['lat'] - min_lat) / dlat) - 1
    if x >= 0 and x <= 99 and y >= 0 and y <= 99:
        Map[x][y] += 1

from sklearn import preprocessing
max_abs_scaler = preprocessing.MaxAbsScaler()
#Map = max_abs_scaler.fit_transform(Map)
data.apply(count, axis = 1)

Map_data = []
for x in range(0, 100):
    for y in range(0, 100):
        Map_data.append([min_lat + dlat*y, min_lng + dlng*x, Map[x][y] - 4 if Map[x][y] - 4 > 0 else 0])
        
from folium.plugins import HeatMap

m = folium.Map([30.508154, 114.418808], zoom_start=14)

HeatMap(Map_data).add_to(m)
#m.save(os.path.join('results', 'C:\\Users\\Wang\\Desktop\\heatp\\Heatmap.html'))

# m.save('C:\\Users\\Wang\\Desktop\\heatp\\Heatmap.html')

m
```
